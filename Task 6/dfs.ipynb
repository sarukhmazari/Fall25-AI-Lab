```python
# without stack
graph={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':['G','H'],
    'E':[],
    'F':['I','K'],
    'G':[],
    'H':['L'],
    'I':[],
    'K':['M'],
    'L':[],
    'M':[]
}

def dfs(graph,node,visited=set()):
    if node not in visited:
        print(node,end=" ")
        visited.add(node)
        for n in graph[node]:
            dfs(graph,n,visited)

dfs(graph,'A')
```

```python
#with stack
graph={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':['G','H'],
    'E':[],
    'F':['I','K'],
    'G':[],
    'H':['L'],
    'I':[],
    'K':['M'],
    'L':[],
    'M':[]
}

def dfs(graph,start):
    visited=set()
    stack=[start]
    while stack:
        s=stack.pop()
        if s not in visited:
            print(s,end=" ")
            visited.add(s)
            stack.extend(reversed(graph[s]))

dfs(graph,'A')
```

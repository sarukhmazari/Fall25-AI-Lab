```python
class Graph:
    def __init__(self, graph):
        self.graph = graph

    def neighbors(self, node):
        return self.graph[node]

    def h(self, node):
        h_values = {'A': 1, 'B': 1, 'C': 1, 'D': 1}
        return h_values[node]

    def a_star(self, start, goal):
        open_nodes = set([start])
        closed_nodes = set([])
        g = {start: 0}
        parent = {start: start}

        while open_nodes:
            current = None

            for node in open_nodes:
                if current is None or g[node] + self.h(node) < g[current] + self.h(current):
                    current = node

            if current is None:
                print("No path found!")
                return None

            if current == goal:
                path = []
                while parent[current] != current:
                    path.append(current)
                    current = parent[current]
                path.append(start)
                path.reverse()
                print("Path found:", path)
                return path

            for (next_node, cost) in self.neighbors(current):
                if next_node not in open_nodes and next_node not in closed_nodes:
                    open_nodes.add(next_node)
                    parent[next_node] = current
                    g[next_node] = g[current] + cost
                else:
                    if g[next_node] > g[current] + cost:
                        g[next_node] = g[current] + cost
                        parent[next_node] = current
                        if next_node in closed_nodes:
                            closed_nodes.remove(next_node)
                            open_nodes.add(next_node)

            open_nodes.remove(current)
            closed_nodes.add(current)

        print("no path found")
        return None


graph = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

g = Graph(graph)
g.a_star('A', 'D')
```

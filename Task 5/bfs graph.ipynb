```python
import collections
def bfs(graph,root):
    visited,queue = set(),collections.deque([root])
    visited.add(root)
    while queue:
        vertex = queue.popleft()
        print(str(vertex) + " ", end="")
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
if __name__ == '__main__':
    graph = {
        'A': ['B', 'C'],
        'B': ['D'],
        'C': ['E', 'F'],
        'D': [],
        'E': [],
        'F': []
    }
print("bfs  order is: ")
bfs(graph, 'A')
```

```python
def dfs(g, s, v=None):
    if v is None:
        v = set()
    v.add(s)
    print(s, end=" ")
    for n in g[s]:
        if n not in v:
            dfs(g, n, v)
g = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['E', 'F'],
    'D': [],
    'E': [],
    'F': []
}
print("dfs order: ")
dfs(g, 'A')

```
